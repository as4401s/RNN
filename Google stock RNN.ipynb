```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('Google_Stock_Price_Train.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
series = df['Open'].values
```

```python
series = series.reshape(-1,1)
series
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sc = MinMaxScaler(feature_range=(0,1))   #sclae all values between zero and 1
train = sc.fit_transform(series).flatten()
```

```python
train
```

creating xtrain and y train

```python
x_train = []
y_train = []
```

```python
#creating a dataset with 60 data and then one output

for i in range (60, len(series)):
    x_train.append(train[i-60:i])
    y_train.append(train[i])
```

```python
x_train
```

```python
y_train
```

```python
x_train, y_train = np.array(x_train), np.array(y_train)
```

```python
x_train
```

```python
#reshaping
#Input shape = 3D tensor with shape (batch_size, timesteps, input_dim)
#x.train.shape 0 gives rows and shpe1 gives number of coloumns

x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
```

```python
x_train
```

```python
x_train.shape
```

```python
import tensorflow as tf

from tensorflow.keras.layers import Dense, LSTM, Input, Dropout
from tensorflow.keras.models import Model
```

```python
i = Input (shape = (x_train.shape[1],1))
x = LSTM (50, return_sequences=True)(i)             #retrun = for adding other lstm layers
x = Dropout(0.2)(x)
x= LSTM (50, return_sequences= True)(x)
x = Dropout(0.2)(x)
x= LSTM (50, return_sequences= True)(x)
x = Dropout(0.2)(x)
x= LSTM (50)(x)
x = Dropout(0.2)(x)
x = Dense(1)(x)                                    #output = 1 as we predict just one stock price

model = Model(i,x)
```

```python
model.compile(
  loss='mse',
  optimizer='adam',
)
```

```python
r = model.fit(x_train, y_train,
  epochs=100, batch_size=32)
```

```python
plt.plot(r.history['loss'], label='loss')
plt.legend()
```
